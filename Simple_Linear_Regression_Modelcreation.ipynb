```python
# Importing the Pandas library to use function to create the dataset  
import pandas as pd
```

```python
# To read the csv file data using pandas read_csv function and assign it to variable called dataset
dataset = pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
# Split and assign the input and output values in the variables
independent = dataset[["YearsExperience"]]
dependent = dataset[["Salary"]]
```

```python
independent
```

```python
dependent
```

```python
#import sklearn libraries along with the model_selection file and then import the train_test_split function
from sklearn.model_selection import train_test_split
# Pass the input,output and split ratio size into  train_test_split function and assign the return values to the 4 respective variables.
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
Y_test
```

```python
#import linearmodel file from sklearn library and import the class LinearRegression
from sklearn.linear_model import LinearRegression
#Assign the class to the var regressornew 
regressornew=LinearRegression()
#Using fit function of regressornew and pass the input and output train set values
#To calculate y = w*X + b
regressornew.fit(X_train,Y_train)
```

```python
LinearRegression()
```

```python
# Assign the calculated Weight and intercep values to the variables
Weight=regressornew.coef_
bias = regressornew.intercept_
# W and b are calculated using the model
```

```python

bias
```

```python
Weight
```

```python
#predict the Salary values
y_pred=regressornew.predict(X_test)
```

```python
y_pred
```

```python
#import metrics file and access r2_score function
from sklearn.metrics import r2_score
#r2_score function is used to calculate the difference using validation method
#If it is near or equal to 1 then good else in and around 0 then bad/ invalid
Err_score=r2_score(Y_test,y_pred)
```

```python
Err_score
#Good score
```

```python
# To save the model import the pickle lib
import pickle
# Random filename to save
filename="finalized_linear_model.sav"
#using the dump function writing the values into new file in the binary format
pickle.dump(regressornew,open(filename,'wb'))
```

```python
#load the saved model to calculate the predict value for given input
loadedmodel=pickle.load(open('finalized_linear_model.sav','rb'))
result=loadedmodel.predict([[16]])# input can be given
```
